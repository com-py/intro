```python
import matplotlib.pyplot as plt, numpy as np
import matplotlib.animation as am
%matplotlib notebook
```

Animate $u(x,t)=\sin(x) \sin (\omega t)$

```python
# may need to run this magic function twice
%matplotlib notebook
def updatefig(*args):               # args[0] = frame = t
    t = 0.1*args[0]
    u = np.sin(x) * np.sin(w*t)
    plot.set_data(x, u)             # update data 

w = 1.0
x = np.linspace(0, 2*np.pi, 101)
fig = plt.figure()
plot = plt.plot(x, 0*x)[0]           # create plot object 
ani = am.FuncAnimation(fig, updatefig, interval=5, blit=True)     # animate 
plt.ylim(-1.2,1.2);
```
