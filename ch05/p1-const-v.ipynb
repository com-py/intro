### Ch05, `p1-const-v`, Motion with constant velocity

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
x0 = 0.0
v = 2.0
dt = 0.1
t = 0.0
tlist = []
xlist = []
for i in range(10):
    tlist.append(t)
    xlist.append(x0)
    x0 = x0 + v*dt
    t = t + dt
```

```python
plt.plot(tlist, xlist)
plt.xlabel('time (s)')
plt.ylabel('x (m)')
plt.show()
```
