### Ch05, `p2-const-a`, Motion with constant acceleration

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
a = 1.0
v = 2.0
x = 0.0
dt = 0.1
t = 0.0
tlist = []
xlist = []
vlist = []
for i in range(20):
    tlist.append(t)
    xlist.append(x)
    vlist.append(v)    
    x = x + v*dt
    v = v + a*dt
    t = t + dt
```

```python
plt.plot(tlist, xlist)
plt.xlabel('time (s)')
plt.ylabel('x (m)')
plt.figure()
plt.plot(tlist, vlist)
plt.xlabel('time (s)')
plt.ylabel('v (m/s)')    
plt.show()
```
