### Ch08, `p1-decay`, Nuclear decay

```python
from numpy.random import random
import matplotlib.pyplot as plt
%matplotlib ipympl
```

```python
p = 0.05
N = 1000
Nlist = []

while N>10:
    Nlist.append(N)
    dN = 0
    for _ in range(N):
        if random() < p:
            dN = dN - 1
    N = N + dN
```

```python
%matplotlib notebook
plt.plot(range(len(Nlist)), Nlist, '.')
plt.xlabel('time (arb.)')
plt.ylabel('$N$')
plt.show()
```
